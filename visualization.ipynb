# Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

# Data Importation

```python
sns.set(style='white')
plt.rcParams['font.family']='cursive'
plt.rcParams['font.size']=30
plt.rc('axes',titlesize=20)
plt.rc('axes',labelsize=20)
plt.rcParams['text.color']='black'
plt.rcParams['axes.labelcolor']='black'
plt.rcParams['xtick.color']='black'
plt.rcParams['ytick.color']='black'
```

```python
df=pd.read_csv('dataset/final_dataset.csv')
df.shape
```

```python
df
```

# Data Preprocessing

```python
df2=df[df['peak time']>df['end time']]
df.drop(df2.index,inplace=True)
df
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df=df.sort_values(by='date')
```

```python
df=df.dropna()
df
```

# Data Exploration

```python
df.describe()
```

```python
df['flare type'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='flare type')
```

```python
df['duration']=df['end time']-df['start time']
df6=df[df['flare type']=='B']
df7=df[df['flare type']=='C']
df8=df[df['flare type']=='M']
df9=df[df['flare type']=='X']
```

```python
w=dict(df['year'].value_counts())
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=w,y=w.values(),x=w.keys(),color='black')
plt.xlabel('Years')
plt.ylabel('Count of flares emitted')
plt.title('Flares emitted from 1981-2017')
```

```python
plt.figure(figsize=(20,7))
plt.bar(df6['year'],df6['duration'],color='white',edgecolor='black')
plt.ylabel('Duration(Sec)')
plt.xlabel('Year')
plt.title('Duration of flares per Year (B Flares)')
```

```python
plt.figure(figsize=(20,7))
plt.bar(df7['year'],df7['duration'],color='white',edgecolor='black')
plt.ylabel('Duration(Sec)')
plt.xlabel('Year')
plt.title('Duration of flares per Year (C Flares)')
```

```python
plt.figure(figsize=(20,7))
plt.bar(df8['year'],df8['duration'],color='white',edgecolor='black')
plt.ylabel('Duration(Sec)')
plt.xlabel('Year')
plt.title('Duration of flares per Year (M Flares)')
```

```python
plt.figure(figsize=(20,7))
plt.bar(df9['year'],df9['duration'],color='white',edgecolor='black')
plt.ylabel('Duration(Sec)')
plt.xlabel('Year')
plt.title('Duration of flares per Year (X Flares)')
```

```python
year_fl= df6[['flare type','year']].groupby(['year']).count().reset_index()
year_fl.columns = ['Year', 'flare count']
dd=df6.groupby(by='year')['Area','DSN','flare type'].sum()
s=list(dd.index)
dd['Year']=s
dd=dd.merge(year_fl)
plt.figure(figsize=(20,6))
plt.plot(dd['Year'],dd['Area'],color='black',linestyle='dashed',linewidth=3)
plt.plot(dd['Year'],dd['DSN'],color='black',linestyle='dashed',linewidth=2)
plt.legend(labels=['Area','Daily sunspot Number','Flare count'])
plt.xlabel('Years')
plt.ylabel('Numbers')
plt.title('Plot to show Sunspot Area, flares and Daily sunspot Number for each Year(B flares)')
```

```python
plt.figure(figsize=(20,6))
plt.bar(dd['Year'],dd['flare count'],color='black')
plt.xlabel('Year')
plt.ylabel('Flare count')
plt.title('Numbers of associated B flares emitted per year')
```
